## **<code>iloc</code>** gets rows (or columns) at particular positions in the index (so it only takes integers).

```python
# df.iloc[rows,columns]
```

```python
df.iloc[:5,2:4]
```

## **<code>loc</code>** gets rows (or columns) with particular labels from the index.

```python
df.loc[:5,['budget','genres']]
```

## **Let's say if you wanted a new column whose values were half the movie runtime**

```python
df.columns
```

```python
df['runtime'].head()
```

```python
df['half_runtime'] = df['runtime']/2
df.head()
```

```python
df.columns
```

```python
df['profit'] = df['revenue'] - df['budget']

df.head()
```

```python
df['profit'].max()
```

```python
df['profit'].min()
```

```python
df['profit'].median()
```

```python
df['profit'].mean()
```

```python
df[df['profit'] == df['profit'].max()][['title','profit']]
```

```python
df[df['profit'] == df['profit'].min()][['title','profit']]
```
