## Python program to calculate the sum of the given series.

1 + x**2/2 + x**3/3 + x**n/n

```python
x = int(input('Enter Your number:'))
n = int(input('Enter nth value'))
sum = 1
for i in range(2,n+1):
    sum = sum + ((x**i)/i)
print(sum)
```
